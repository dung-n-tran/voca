```python
train_subjects, val_subjects, test_subjects = pickle.load(open('../split.pkl', 'rb'))
template_dir = '/media/gaurav/DATADRIVE0/datasets/voca/templates/'
data_dir = '/media/gaurav/DATADRIVE0/datasets/voca/unposeddata//'
count = 0
all_face_diffs = np.zeros((123710, 5023, 3))
for t in tqdm(plb.Path(template_dir).iterdir(), total=12):
    if not str(t).endswith('pdf') and t.name in train_subjects:
        print(t.name)
        template_face = Mesh(filename=str(t))
        face_dir = os.path.join(data_dir, t.name.split('.ply')[0])
        for sent in plb.Path(face_dir).iterdir():
            print(sent.name)
            for ply in sent.iterdir():
                if str(ply).endswith('.ply'):
                    inst_face = Mesh(filename=str(ply))
                    diff = inst_face.v - template_face.v
                    all_face_diffs[count] = diff
                    count += 1
```

```python
count
```

```python
for_pca = all_face_diffs[:count]
```

```python
for_pca.shape
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=50)
```

```python
pca_comps = pca.fit(for_pca.reshape(count, 5023*3).T)
```

```python
pca_comps = pca.transform(for_pca.reshape(count, 5023*3).T)
```

```python
pickle.dump(pca_comps, open('../train_pca_components_50_unposed.pkl', 'wb'))
```

```python
ds_dir = '/media/gaurav/DATADRIVE0/datasets/voca/ds_features/'
audio_dir = '/media/gaurav/DATADRIVE0/datasets/voca/audio/'
config = {}
config['deepspeech_graph_fname'] = '../ds_graph/models/output_graph.pb'
config['audio_feature_type'] = 'deepspeech'
config['num_audio_features'] = 29

config['audio_window_size'] = 16
config['audio_window_stride'] = 1
audio_handler = AudioHandler(config)
tmp_audio = {}
for s in plb.Path(audio_dir).iterdir():
    tmp_audio[s.name] = {}
    for a in s.iterdir():
        print(str(a))
        sample_rate, audio = wavfile.read(str(a))
        tmp_audio[s.name][a.name] = {'audio': audio, 'sample_rate': sample_rate}
        
x = audio_handler.process(tmp_audio)
for subj in x.keys():
    if not os.path.exists(os.path.join(ds_dir, subj)):
        os.makedirs(os.path.join(ds_dir, subj))
    for seq in x[subj].keys():
        print(os.path.join(ds_dir, subj, seq.split('.wav')[0]+'.npy'))
        np.save(os.path.join(ds_dir, subj, seq.split('.wav')[0]+'.npy'), x[subj][seq]['audio'])
```

# splits

```python
train, val, test = pickle.load(open('../split.pkl', 'rb'))
```

```python
train_data = []
ds_dir = '/media/gaurav/DATADRIVE0/datasets/voca/ds_features/'
ply_dir = '/media/gaurav/DATADRIVE0/datasets/voca/unposeddata//'
for t in train:
    sp = t.split('.ply')[0]
    for sentence_npy in plb.Path(os.path.join(ds_dir, sp)).iterdir():
        speech_feat = np.load(str(sentence_npy))
        sentence = sentence_npy.name.split('.npy')[0]
        print(speech_feat.shape)
        for i in range(1, speech_feat.shape[0]+1):
            path = os.path.join(ply_dir, sp, sentence, '{}.{:06d}.ply'.format(sentence, i))
#             print(path)
            if os.path.exists(path):
                train_data.append('{}/{}/{}'.format(sp, sentence, i))
            else:
                train_data.pop()
                break
        if i == speech_feat.shape[0]:
            train_data.pop()
#         print(train_data)
        print(len(train_data))
#         break
#     break
```

```python
with open('../splits/train.txt', 'w') as f:
    for t in train_data:
        f.write(t+'\n')
```

```python
val_data = []
ds_dir = '/media/gaurav/DATADRIVE0/datasets/voca/ds_features/'
ply_dir = '/media/gaurav/DATADRIVE0/datasets/voca/registereddata/'
for t in val:
    sp = t.split('.ply')[0]
    for sentence_npy in plb.Path(os.path.join(ds_dir, sp)).iterdir():
        speech_feat = np.load(str(sentence_npy))
        sentence = sentence_npy.name.split('.npy')[0]
        print(speech_feat.shape)
        for i in range(1, speech_feat.shape[0]+1):
            path = os.path.join(ply_dir, sp, sentence, '{}.{:06d}.ply'.format(sentence, i))
#             print(path)
            if os.path.exists(path):
                val_data.append('{}/{}/{}'.format(sp, sentence, i))
            else:
                val_data.pop()
                break
        if i == speech_feat.shape[0]:
            val_data.pop()
#         print(val_data)
        print(len(val_data))
#         break
#     break
```

```python
with open('../splits/val.txt', 'w') as f:
    for t in val_data:
        f.write(t+'\n')
```

```python
splits = {'train':[t.split('.ply')[0] for t in train], 'val':[t.split('.ply')[0] for t in val], 'test':[t.split('.ply')[0] for t in test]}
print(splits)
pickle.dump(splits, open('../split_dict.pkl', 'wb'))
```
